# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.info()
```

# Histogram plot

```python
plt.Figure(figsize=(8,5))

sns.histplot(
    data = df,
    x = "age",
    bins=20,
    kde=True,
    color="blue"
)
plt.title("Age Distribution")
plt.show()
```

# KDE Plot

```python
sns.kdeplot(
    data=df,
    x="age",
    fill=True
)
plt.show()
```

# Scatter plot

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x="age",
    y="fare",
    hue="survived",
    style="sex",
    s=100,
)
plt.show()
```

# line

```python
avg_age = df.groupby("pclass")['age'].mean().reset_index()

sns.lineplot(
    data=avg_age,
    x='pclass',
    y='age',
    marker='o'
)
plt.show()
```

# Box plot

```python
sns.boxplot(
    data=df,
    x="sex",
    y="fare"
)
plt.show()
```

# Viollineplot

data distribution and outlier detect

```python
sns.violinplot(
    data=df,
    x="sex",
    y="fare"
)
plt.show()
```

# pairplot

```python
sns.pairplot(
    df[
        [
            "age",
            "fare",
            "sibsp",
            "parch",
            "survived"
        ]
    ],
    hue="survived"
)
plt.show()
```

```python
corr = df.select_dtypes("number").corr()

plt.Figure(figsize=(8,6))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)
```
