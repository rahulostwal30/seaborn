```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display,HTML


class Hotel_booking:
    
    def header(self,header,font=("arial",18,"bold")):
        display(HTML(f"<h2 style='text-align:center;color:blue'>{header} </h2>"))

    def Dataset_understanding(self,data):
        self.header("First Five rows")
        display(data.head())
        self.header("Find")
        display("Number of Rows")
        display(data.shape[0])
        display("Number of columns")
        display(data.shape[1])
        display("Data type of all columns")
        display(data.info())
        self.header("Missing values")
        display(data.isna().sum())
        self.header("Duplicate records")
        display(data.duplicated())
        self.header("Statistical summary of all numerical columns")
        display(data.describe())
        self.header("List all features")
        display(data.columns.to_list())

    def Data_visualization(self,data):
        self.header("Histogram")
        plt.Figure(figsize=(6,4))
        sns.histplot(
            data=data,
            x="adr",
            bins=30,
            kde=True,
            color="Red"
        )
        plt.title("Adr Distribution")
        plt.show()
        print("""
            1. Is the distribution normally distributed?\n
            - The ADR distribution is not normally distributed.\n
            2. Is it positively or negatively skewed?\n
            - It is positively right skewed.\n
            3. Are there any unusually high room prices?\n
            - yes"""
        )
        #--------------------------------------
        self.header("KDE Plot")
        plt.figure(figsize=(5,3))
        sns.kdeplot(
            data=data,
            x="lead_time",
            fill=True,
        )
        plt.title("Lead Time Dinsity")
        plt.show()
        print("""
                1. At approximately what lead time is the booking density highest?\n
                The booking density is highest at approximately 0–50 days of lead time\n
                2. Do most customers book well in advance or close to the check-in date?\n
                Most customers book relatively close to the check-in date rather than several months in advance\n
            """)
        # ------------------------------------
        self.header("Scatter Plot")
        plt.Figure(figsize=(6,4))
        sns.scatterplot(
            data=data, 
            x="lead_time",
            y="adr",
            hue="is_canceled",
            style="hotel",
            size="total_of_special_requests",
            s=100
        )
        plt.legend()
        plt.title("ADR vs Lead Time")
        plt.show()
        print("""1. Is there any relationship between Lead Time and ADR?\n
                    There is no strong linear relationship between Lead Time and ADR\n
                    2. Which type of bookings are canceled more frequently?\n
                    Bookings with longer lead times tend to have higher cancellation rates.\n
                    3. Do customers requesting more special services tend to cancel less?\n
                    Yes. Customers who make more special requests generally appear to cancel less frequently.
                """)
        #-------------------------------------
        self.header("Box plot")
        sns.scatterplot(
            data=data,
            x="hotel",
            y="adr"
        )
        plt.legend()
        plt.show()
        print("""1. Which hotel type has the higher average room price?\n
                City Hotel has the higher average ADR compared to Resort Hotel\n
                2. Which hotel contains more outliers?\n
                City Hotel contains more outliers\n
                3. What does the median line represent?\n
                The median line represents the middle value (50th percentile) of the data. It divides the data into two equal parts""")
        #---------------------------------------------
        self.header("violin plot")
        sns.violinplot(
            data=data,
            x="lead_time",
            y="customer_type"
        )
        plt.show()
        print("""
                1. Which customer type books the earliest?\n
                Transient customers book the earliest.\n
                2. Which customer type shows the highest booking cancellations?\n
                Transient customers show the highest booking cancellations.\n
                3. How is a violin plot different from a box plot?\n
                violin plot show the distribution of the data & box plot is used to outlier detection 
        """)
        #---------------------------
        self.header("count plot")
        sns.countplot(data=data,
                      x="market_segment"
                      )
        plt.xticks(rotation=90)
        plt.show()
        print(""" 
                1. Which market segment contributes the highest number of bookings?\n
                Online TA has the highest number of bookings.\n
                2. Which market segment has the highest number of cancellations?\n
                Online TA has the highest number of cancellations.
            """)
        #-------------------------------
        self.header("Bar plot")
        sns.barplot(
            data=data,
            x="meal",
            y="adr"
        )
        plt.title("meal vs adr")
        plt.show()
        #-------------------------
        self.header("correlation Heatmap")
        corr = data.select_dtypes("number").corr()
        plt.Figure(figsize=(6,5))
        sns.heatmap(
            corr,
            annot=True,
            cmap='coolwarm'
        )
        plt.show()
        print("""
                1. Which two features have the strongest positive correlation?\n
                stays_in_weekend_nights and stays_in_week_nights\n
                2. Which feature has the strongest negative correlation?\n
                is_canceled and total_of_special_requests\n
                3. Which numerical features appear to have very little relationship?\n
                adr and lead_time\n
                adr and days_in_waiting_list\n
                babies and lead_time\n""")
        #---------------------------------------------
        self.header("Pairplot")
        sns.pairplot(
            data[["lead_time","adr","adults","children","total_of_special_requests","is_canceled"]],
            hue="is_canceled"
        )
        plt.show()
        print("""
            1. Which variables appear to be positively correlated?\n
            stays_in_week_nights and stays_in_weekend_nights\n
            2. Which variables show little or no relationship?\n
            adr and (lead_time,days_in_wating) & babies and lead_time\n
            3. Which feature best distinguishes canceled and non-canceled bookings?\n
            lead_time
        """)

    def bussiness_insights(self,data):
        pass



data = pd.read_csv("hotel_bookings.csv")
obj = Hotel_booking()
```
